```python
movie_matrix=data.pivot_table(index='user_id',columns='movie_title',values='rating')
```

```python
movie_matrix.head()
```

```python
titles[titles.index==50]
```

```python
star_wars_ratings=movie_matrix['Star Wars (1977)']
```

```python
similar_to_star_wars=movie_matrix.corrwith(star_wars_ratings)  
# Here NaN values means their is no user in common who watched both
```

```python
corr_star_wars=pd.DataFrame(similar_to_star_wars,columns=['Correlation'])
```

```python
corr_star_wars.dropna(inplace=True)
```

```python
corr_star_wars
# The perfect correlation is because say 6 people watche man of the year and rated 5 and sa,e watched star wars and rated 5.
```

```python
ids_to_use_for_star_wars=ratings[ratings.no_of_times>100].index
```

```python
movies_to_use_for_star_wars=titles.loc[ids_to_use_for_star_wars]
```

```python
movies_to_use_for_star_wars
```

```python
# i have opted the longer path if it works go with it corr_star_wars.join(ratings['no_of_times'])
correlation_df=corr_star_wars.loc[movies_to_use_for_star_wars.movie_title]
```

```python
correlation_df.sort_values('Correlation',ascending=False).head()
```

```python
def recommend_movies(movie_name):
    movie_user_rating=movie_matrix[movie_name]
    similar_to_movies=movie_matrix.corrwith(movie_user_rating)
    corr_of_movie=pd.DataFrame(similar_to_movies,columns=['correlations'])
    corr_of_movie.dropna(inplace=True)
    correlation_data=corr_of_movie.loc[movies_to_use_for_star_wars.movie_title]
    recommend=correlation_data.sort_values('correlations',ascending=False).head()
    return recommend
```

```python
recommend_movies('Titanic (1997)')
```
